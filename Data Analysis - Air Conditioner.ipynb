```python
from pyspark.sql.functions import *
import urllib
from pyspark.sql.functions import monotonically_increasing_id
import pyspark.sql.functions as f
```

```python
display(dbutils.fs.ls("/FileStore/tables/amazon_databricks_project_accessKeys.csv"))
```

```python
header = "true"
inferSchema = "true"
seperator = ","
aws_keys_df = spark.read.format("csv").option("header", header).option ("inferSchema", inferSchema).option("sep", seperator).load("/FileStore/tables/amazon_databricks_project_accessKeys.csv")
```

```python
ACCESS_KEY = aws_keys_df.select("Access key ID").collect()[0]["Access key ID"]
SECRET_KEY = aws_keys_df.select("Secret access key").collect()[0]["Secret access key"]

ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe="")
```

```python
AWS_S3_BUCKET = "aws-airconditioner-project/amazon"
MOUNT_NAME = "/mnt/aws-airconditioner-project"
SOURCE_URL = "s3a://{0}:{1}@{2}".format(ACCESS_KEY, ENCODED_SECRET_KEY, AWS_S3_BUCKET)

dbutils.fs.mount(SOURCE_URL, MOUNT_NAME)
```

```python
display(dbutils.fs.ls("/mnt/aws-airconditioner-project"))
```

```python
FILE_LOCATION = "/mnt/aws-airconditioner-project/Air_Conditioners.csv"
air_conditioner_df = spark.read.format("csv").option("header", header).option("sep", seperator).option("inferSchema", inferSchema).load(FILE_LOCATION)
```

```python
display(air_conditioner_df)
```

```python
air_conditioner_df.write.mode("overwrite").saveAsTable("airconditioner")
```

```python
air_conditioner_df = air_conditioner_df.withColumn("ID", monotonically_increasing_id())
```

```python
air_conditioner_df.write.format("csv").option("header", header).save(
    "/mnt/aws-airconditioner-project/Updated_Air_Conditioners.csv"
)
```

```python
# Find any NULL values exist in any of the columns
air_conditioner_df_Columns = ["Name", "Main_Category", "Sub_Category", "Image", "Link", "Ratings", "No_Of_Ratings",
                              "Discount_Price","Actual_Price", "ID",]
air_conditioner_df.select([count(when(col(c).contains('None') | \
                            col(c).contains('NULL') | \
                            (col(c) == '' ) | \
                            col(c).isNull() | \
                            isnan(c), c 
                           )).alias(c)
                           for c in air_conditioner_df_Columns]).show()
```

```python
# Remove the Special Character "₹$," from the integer
columns_to_translate = ["No_Of_Ratings", "Discount_Price", "Actual_Price"]
for c in columns_to_translate:
    print(c)
    air_conditioner_df = air_conditioner_df.withColumn(c, f.translate(f.col(c), "₹$,", "").alias("replaced"))
air_conditioner_df.show()
```

```python
# Convert the "No_Of_Ratings", "Disount Price" & "Actual Price" into Integer
columns_to_cast = ["No_Of_Ratings", "Discount_Price", "Actual_Price"]
air_conditioner_df = (
   air_conditioner_df
   .select(
     *(c for c in air_conditioner_df.columns if c not in columns_to_cast),
     *(col(c).cast("int").alias(c) for c in columns_to_cast)
   )
)

air_conditioner_df.printSchema()
```

```python
# Convert the  "Ratings" field into "Float"

air_conditioner_df = air_conditioner_df.withColumn(
    "ratings", col("ratings").cast("float"))

air_conditioner_df.printSchema()
```

```python
# Replace NULL values with 0
air_conditioner_df = air_conditioner_df.fillna(0, subset=["Ratings", "No_Of_Ratings", "Discount_Price", "Actual_Price"])
```

```python
# Check if there's any other NULL values in the columns
air_conditioner_df.select([count(when(col(c).contains('None') | \
                            col(c).contains('NULL') | \
                            (col(c) == '' ) | \
                            col(c).isNull() | \
                            isnan(c), c 
                           )).alias(c)
                           for c in air_conditioner_df_Columns]).show()
```

```python
air_conditioner_df.write.option("header", header).mode("overwrite").csv(
    "/mnt/aws-airconditioner-project/Updated_Air_Conditioners.csv"
)
```
